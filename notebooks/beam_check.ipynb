```python
%matplotlib inline

import numpy as np
import pylab as plt
from scipy.ndimage.filters import gaussian_filter
```

```python
d1 = np.random.normal(0,1,(400,400))
d2 = np.random.normal(0,1,(400,400))
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(ncols=3,nrows=1,figsize=(10,5))

ax1.imshow(d1)
ax2.imshow(d2)
ax3.imshow(d1+d2)
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(ncols=3,nrows=1,figsize=(10,5))

ax1.imshow(gaussian_filter(d1,2))
ax2.imshow(gaussian_filter(d2,2))
ax3.imshow(gaussian_filter(d1+d2,2))
```

```python
ss = gaussian_filter(d1+d2,2)-(gaussian_filter(d1,2)+gaussian_filter(d2,2))
ss.sum()
```

```python
from skimage.draw import polygon,circle

d1 = np.zeros((400, 400), dtype=np.uint8)
r = np.array([100, 150, 200])
c = np.array([100,200,100])
rr, cc = polygon(r, c)
d1[rr, cc] = 1

d2 = np.zeros((400, 400), dtype=np.uint8)
rr, cc = circle(100, 250, 100)
d2[rr, cc] = 1
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(ncols=3,nrows=1,figsize=(10,5))

ax1.imshow(gaussian_filter(d1,3))
ax2.imshow(gaussian_filter(d2,3))
ax3.imshow(gaussian_filter(d1+d2,3))
```

```python
ss = gaussian_filter(d1+d2,2)-(gaussian_filter(d1,2)+gaussian_filter(d2,2))
ss.sum()
```
